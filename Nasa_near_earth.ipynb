# Collecting 10k records

```python
import requests
import pandas as pd


API_KEY = "dZrRqJahDPm6OWjfavosphf6C8hT9AZJs9bXiMva"


start_url = f"https://api.nasa.gov/neo/rest/v1/feed?start_date=2024-01-01&end_date=2024-01-07&api_key={API_KEY}"

all_asteroids = []
row_limit = 10000

while start_url and len(all_asteroids) < row_limit:
    response = requests.get(start_url)

    data = response.json()
    neo_data = data['near_earth_objects']

    for date, asteroids in neo_data.items():
        for asteroid in asteroids:
            asteroid_data = {
                'id': asteroid['id'],
                'name': asteroid['name'],
                'absolute_magnitude_h': asteroid['absolute_magnitude_h'],
                'est_diameter_min_km': asteroid['estimated_diameter']['kilometers']['estimated_diameter_min'],
                'est_diameter_max_km': asteroid['estimated_diameter']['kilometers']['estimated_diameter_max'],
                'is_potentially_hazardous': asteroid['is_potentially_hazardous_asteroid'],
                'close_approach_date': asteroid['close_approach_data'][0]['close_approach_date'],
                'relative_velocity(km/h)':asteroid['close_approach_data'][0]['relative_velocity']['kilometers_per_hour'],
                'astronomical_unit':asteroid['close_approach_data'][0]['miss_distance']['astronomical'],
                'lunar_distance':asteroid['close_approach_data'][0]['miss_distance']['lunar'],
                'miss_distance_km':asteroid['close_approach_data'][0]['miss_distance']['kilometers'],
                'orbiting_body':asteroid['close_approach_data'][0]['orbiting_body']
            }
            all_asteroids.append(asteroid_data)

            # Stop if we reach the row limit
            if len(all_asteroids) >= row_limit:
                break
        if len(all_asteroids) >= row_limit:
            break

    start_url = data['links'].get('next')
```

# Asteroids table

```python
cursor.execute("""create table if not exists asteroids(
  id INT,
  name VARCHAR(100),
  absolute_magnitude_h FLOAT,
  estimated_diameter_min_km FLOAT,
  estimated_diameter_max_km FLOAT,
  is_potentially_hazardous_asteroid BOOL
)
""")
```

```python
insert_query = "insert into asteroids values(?,?,?,?,?,?)"

for i in all_asteroids:
  id = int(i['id'])
  name = i['name']
  magnitude = float(i['absolute_magnitude_h'])
  dia_min = float(i['est_diameter_min_km'])
  dia_max = float(i['est_diameter_max_km'])
  hazard = i['is_potentially_hazardous']

  values = (id,name,magnitude,dia_min,dia_max,hazard)

  cursor.execute(insert_query,values)
conn.commit()
```

# Close Approach table

```python
cursor.execute("""create table if not exists close_approach(
  neo_reference_id INT,
  close_approach_date DATE,
  relative_velocity_kmph FLOAT,
  astronomical FLOAT,
  miss_distance_km FLOAT,
  miss_distance_lunar FLOAT,
  orbiting_body text)""")
```

```python
from datetime import datetime

insert_query = 'insert into close_approach values(?,?,?,?,?,?,?)'

for i in all_asteroids:
      id = int(i['id'])
      formatted_date = datetime.strptime(i['close_approach_date'], "%Y-%m-%d").date()
      velocity = float(i['relative_velocity(km/h)'])
      astro= float(i['astronomical_unit'])
      miss_dist= float(i['miss_distance_km'])
      lunar= float(i['lunar_distance'])
      body = i['orbiting_body']

      values = (id,formatted_date,velocity,astro,miss_dist,lunar,body)

      cursor.execute(insert_query,values)

conn.commit()

```

# Queries

#1.Count how many times each asteroid has approached Earth

```python
cursor.execute("""
    SELECT a.name, COUNT(*) AS approach_count
    FROM close_approach ca
    JOIN asteroids a
      ON a.id = ca.neo_reference_id
    WHERE ca.orbiting_body = 'Earth'
    GROUP BY a.name
    ORDER BY approach_count DESC
""")

result = cursor.fetchall()

# Put into a DataFrame
data = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data
```

# 2. Average velocity of each asteroid over multiple approaches

```python
cursor.execute("""
    SELECT a.name,
           AVG(ca.relative_velocity_kmph) AS avg_velocity_kmph,
           COUNT(*) AS total_approaches
    FROM close_approach ca
    JOIN asteroids a
      ON a.id = ca.neo_reference_id
    GROUP BY a.name
    ORDER BY avg_velocity_kmph DESC
""")

result = cursor.fetchall()

# Put into DataFrame
data = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data['avg_velocity_kmph'] = data['avg_velocity_kmph'].astype(float).round(2)
data
```

# List top 10 fastest asteroids

```python
cursor.execute("""SELECT name,relative_velocity_kmph FROM close_approach
                join asteroids
               on asteroids.id = close_approach.neo_reference_id
               order by relative_velocity_kmph desc
               limit 10""")
result = cursor.fetchall()
data = pd.DataFrame(result,columns = [i[0] for i in cursor.description])
data
```

4. Find potentially hazardous asteroids that have approached Earth more than 3 times

```python
cursor.execute("""
    SELECT a.name, COUNT(*) AS earth_approach_count
    FROM close_approach ca
    JOIN asteroids a
      ON a.id = ca.neo_reference_id
    WHERE ca.orbiting_body = 'Earth'
      AND a.is_potentially_hazardous_asteroid = 1   -- assuming 1 = True
    GROUP BY a.name
    HAVING COUNT(*) > 3
    ORDER BY earth_approach_count DESC
""")

result = cursor.fetchall()

data = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data
```

# Find the month with the most asteroid approaches

```python
cursor.execute("""
    SELECT
        strftime('%m', close_approach_date) AS month,
        COUNT(*) AS approach_count
    FROM
        close_approach
    GROUP BY
        month
    ORDER BY
        approach_count DESC
    LIMIT 1;
""")

result = cursor.fetchone()
print("Most frequent month (MM):", result)
```

9. Display the name of each asteroid along with the date and miss distance of its closest approach to Earth.

```python
cursor.execute("""
    SELECT a.name,
           ca.close_approach_date,
           ca.miss_distance_km
    FROM close_approach ca
    JOIN asteroids a
      ON a.id = ca.neo_reference_id
    WHERE ca.orbiting_body = 'Earth'
      AND ca.miss_distance_km = (
          SELECT MIN(ca2.miss_distance_km)
          FROM close_approach ca2
          WHERE ca2.neo_reference_id = ca.neo_reference_id
            AND ca2.orbiting_body = 'Earth'
      )
    ORDER BY ca.miss_distance_km ASC
""")

result = cursor.fetchall()

data = pd.DataFrame(result, columns=[i[0] for i in cursor.description])
data
```
